## 2. Extract the clinically useful information from the whole data

The whole data are saved as follows:

* Registered data follow each other 
* At regular intervals, the displayed value is 254 which actually corresponds to a signal tag.  
* Then at this tag value + 1, the displayed value is the percentage of functional oxygen saturation 
* Then at this tag value + 2, the displayed value is the pulse rate

Example :
```python
...
30  : data  
254 : tag value t, no corresponding data  
97  : t+1 --> value of the percentage of functional oxygen saturation
72  : t+2 --> value of the pulse rate  
26  : data  
27  : data  
26  : data  
...
```

````python
data = df.values#[0:300]

# List of tag index values
tagindex = np.where(data == 254)[0]

print("List of tag index values:", tagindex)
# Check if for the first tag index value, the signal value is indeed 254
#print(data[tagindex[0]][0])

# Define a set of empty lists to store the clinically useful information
sat_arr = []
fc_arr = []
data_arr = data.copy()
valuestoremove = []

fc = np.zeros(data.size)
fc2 = np.zeros(data.size)
fc[:] = np.nan
fc2[:] = np.nan

sat = np.zeros(data.size)
sat2 = np.zeros(data.size)
sat[:] = np.nan
sat2[:] = np.nan
````

````python
for i in range(tagindex.size):
    # Store value of saturation for each tag index
    sat_arr.append(data[tagindex[i]+1][0])
    # Store value of pulse rate for each tag index
    fc_arr.append(data[tagindex[i]+2][0])
    # Store corresponding indices as well as tag and closing values ones
    valuestoremove.append([tagindex[i],tagindex[i]+1,tagindex[i]+2])
    fc[tagindex[i]]=fc_arr[i]
    fc2[tagindex[i]:]=fc_arr[i]
    sat[tagindex[i]]=sat_arr[i]
    sat2[tagindex[i]:]=sat_arr[i]

# Remove these values from the original data
data_arr=np.delete(data_arr,valuestoremove)
````

## 3. Plot the results

````python
plt.subplots(2,2,figsize=(15,12))

plt.subplot(3,2,1)
plt.plot(data)
plt.ylim(0,255)
plt.title('Complete oxypleth data')

plt.subplot(3,2,2)
plt.plot(data_arr[0:300])
plt.ylim(0,255)
plt.title('Filtered oxypleth data')

plt.subplot(3,2,3)
plt.plot(range(fc.size),fc,'o')
plt.plot(range(sat.size),sat,'ro')
plt.legend(['Pulse rate','Saturation'])
plt.ylim(0,120)

plt.subplot(3,2,4)
plt.plot(data_arr[0:300],'-g')
plt.title('Filtered oxypleth data')

plt.subplot(3,2,5)
plt.plot(fc2,'o')
plt.legend(['Pulse rate'])

plt.subplot(3,2,6)
plt.plot(sat2,'or')
plt.legend(['Saturation'])

plt.show()
````

````python
# get the lowest value (except nan) of the heartrate
fc_min = (np.nanmin(fc))
#index_fc_min = (fc.index(np.nanmin(fc)))

index_fc_min = (list(fc).index(np.nanmin(fc)))

fc_max = (np.nanmax(fc))
index_fc_max = (list(fc).index(np.nanmax(fc)))

print("Fc min : ",fc_min)
print("Fc max : ",fc_max)
````

````python
# final result and save
save_flag = True # if True, 2 images will be saved in the output folder

%matplotlib inline
plt.subplots(2,1,figsize=(10,8))

plt.subplot(2,1,1)
plt.plot(data_arr[700:1000])
plt.ylim(0,100)
plt.title('Signal')

plt.subplot(2,1,2)
#plt.ylim(0,100)
plt.ylim(ymax=70)
p1=plt.plot(sat2,"--",label="saturation")
p2=plt.plot(fc2,"r",label="heartrate")
plt.ylim(0,110)
plt.text(index_fc_max,fc_max-10, "maximum", horizontalalignment='left',family="sans-serif")

plt.annotate('hr minimum', xy=(index_fc_min-2,fc_min-2), xytext=(500, 40), horizontalalignment='center',
            arrowprops=dict(arrowstyle="->", facecolor='black'))


plt.legend(loc="lower right", \
           ncol=1, shadow=True, title="Legend", fancybox=True, framealpha=0) # alpha 1=opaque 0=transparent
plt.title("Saturation and Hr")
# sauvegarde dans des fichiers images
if save_flag:
    plt.savefig(os.path.join(OUTPUT_DIR,'test.png'),dpi=150,transparent=True) # try different values of dpi, other format like jpg and tiff, zoom on the curve
# discuss between jpg format (good for picture, to avoid if curves, ie destructive compression) and png format (non destructive, alpha canal)
    plt.savefig(os.path.join(OUTPUT_DIR,'test.svg')) # try the svg and pdf format
# discuss about the different (size, fidelity) of the vectoriel format agains bitmap format
````
