```python
import numpy as np
import matplotlib.pyplot as plt


A = 0.1
B = 0.2
t = np.arange(0,100,dtype=float)
Vp = 100
V = Vp * 2.334 * (np.exp(-A*t)-np.exp(-B*t))
plt.plot(t,V)
```

```python
V2 = 60 / (1+60*np.exp(-0.8*t)) 
plt.plot(t,V2)
```

```python
V3 = V+V2
plt.plot(t,V3)
```
