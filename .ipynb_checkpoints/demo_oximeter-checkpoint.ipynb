## 2. Extract the clinically useful information from the whole data

The whole data are saved as follows:

* Registered data follow each other 
* At regular intervals, the displayed value is 254 which actually corresponds to a signal tag.  
* Then at this tag value + 2, the displayed value is the percentage of functional oxygen saturation 
* Then at this tag value + 4, the displayed value is the pulse rate
* Then at this tag + 6, the displayed value is 0

Example :
```python
...
30  : data  
254 : tag value t, no corresponding data 
27  : data  
97  : t+2 --> value of the percentage of functional oxygen saturation
26  : data
72  : t+4 --> value of the pulse rate  
26  : data  
0   : t+6 --> closing value for saturation pulse rate information, no data  
27  : data  
26  : data  
...
```

````python
data = df.values
# List of tag index values
tagindex = np.where(df == 254)[0]
print("List of tag index values:", tagindex)
# Check if for the first tag index value, the signal value is indeed 254
print(data[tagindex[0]][0])

# Define a set of empty lists to store the clinically useful information
sat_arr = []
fc_arr = []
data_arr = data.copy()
valuestoremove = []

fc = np.zeros(df.size)
fc2 = np.zeros(df.size)
fc[:] = np.nan
fc2[:] = np.nan

sat = np.zeros(df.size)
sat2 = np.zeros(df.size)
sat[:] = np.nan
sat2[:] = np.nan
````

````python
for i in range(tagindex.size):
    # Store value of saturation for each tag index
    sat_arr.append(data[tagindex[i]+2][0])
    # Store value of pulse rate for each tag index
    fc_arr.append(data[tagindex[i]+4][0])
    # Store corresponding indices as well as tag and closing values ones
    valuestoremove.append([tagindex[i],tagindex[i]+2,tagindex[i]+4,tagindex[i]+6])
    fc[tagindex[i]]=fc_arr[i]
    fc2[tagindex[i]:]=fc_arr[i]
    sat[tagindex[i]]=sat_arr[i]
    sat2[tagindex[i]:]=sat_arr[i]

# Remove these values from the original data
data_arr=np.delete(data_arr,valuestoremove)
````

## 3. Plot the results

````python
plt.subplots(2,2,figsize=(15,12))

plt.subplot(3,2,1)
plt.plot(data)
plt.ylim(0,255)
plt.title('Complete oxypleth data')

plt.subplot(3,2,2)
plt.plot(data_arr)
plt.ylim(0,255)
plt.title('Filtered oxypleth data')

plt.subplot(3,2,3)
plt.plot(range(fc.size),fc,'o')
plt.plot(range(sat.size),sat,'ro')
plt.legend(['Pulse rate','Saturation'])
plt.ylim(0,120)

plt.subplot(3,2,4)
plt.plot(data_arr,'-g')
plt.title('Filtered oxypleth data')

plt.subplot(3,2,5)
plt.plot(fc2,'o')
plt.legend(['Pulse rate'])

plt.subplot(3,2,6)
plt.plot(sat2,'or')
plt.legend(['Saturation'])

plt.show()
````
